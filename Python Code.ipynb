```python
# Importing necessary libraries
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import numpy as np


# Initializing the Random Forest Regressor model
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)  # You can adjust n_estimators as needed

# Training the model on the training data
rf_regressor.fit(X_train, y_train)
rf_regressor.fit(X_train_scaled, y_train_scaled)

# Predicting on the training set
train_predictions = rf_regressor.predict(X_train)
train_predictions_scaled = rf_regressor.predict(X_train_scaled)

# Predicting on the test set
test_predictions = rf_regressor.predict(X_test)
test_predictions_scaled = rf_regressor.predict(X_test_scaled)

# Evaluating the model
train_rmse = np.sqrt(mean_squared_error(y_train, train_predictions))
train_rmse_scaled = np.sqrt(mean_squared_error(y_train_scaled, train_predictions_scaled))
test_rmse = np.sqrt(mean_squared_error(y_test, test_predictions))
test_rmse_scaled = np.sqrt(mean_squared_error(y_test_scaled, test_predictions_scaled))

print("Train RMSE:", train_rmse)
print("Train RMSE scaled:", train_rmse_scaled)

print("Test RMSE:", test_rmse)
print("Test RMSE scaled:", test_rmse_scaled)
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid to search through
param_grid = {
    'n_estimators': [100, 200, 300],  # Number of trees in the forest
    'max_depth': [None, 10, 20, 30],   # Maximum depth of the trees
    'min_samples_split': [2, 5, 10],   # Minimum number of samples required to split an internal node
    'min_samples_leaf': [1, 2, 4]      # Minimum number of samples required to be at a leaf node
}

# Initialize the Random Forest Regressor model
rf_regressor = RandomForestRegressor(random_state=42)

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=rf_regressor, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error')
# Convert DataFrame to numpy array and then flatten
y_train = y_train.to_numpy().ravel()
y_test = y_test.to_numpy().ravel()


# Now, the target variables are reshaped to 1D arrays

# Fit the grid search to the data
grid_search.fit(X_train, y_train)

# Get the best parameters and the best score
best_params = grid_search.best_params_
best_score = np.sqrt(-grid_search.best_score_)  # We take the square root of the negative mean squared error to get RMSE

print("Best Parameters:", best_params)
print("Best RMSE:", best_score)

# Predict using the best model found by GridSearchCV
best_model = grid_search.best_estimator_
test_predictions = best_model.predict(X_test)
test_rmse = np.sqrt(mean_squared_error(y_test, test_predictions))
print("Test RMSE with Best Model:", test_rmse)
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error

param_grid = {
    'hidden_layer_sizes': [(50, 50), (100, 100), (50,)],
    'activation': ['relu', 'tanh'],
    'solver': ['adam', 'lbfgs', 'sgd'],
    'max_iter': [9000,9500,10000]
}


# Create MLPRegressor
mlp_regressor = MLPRegressor(random_state=42)

# Instantiate GridSearchCV
grid_search = GridSearchCV(mlp_regressor, param_grid, cv=5, scoring='neg_mean_squared_error')

# Perform GridSearchCV
grid_search.fit(X_train_scaled, y_train)

# Best parameters
print("Best Parameters:", grid_search.best_params_)

# Best estimator
best_model = grid_search.best_estimator_

# Predictions on test set
y_pred = best_model.predict(X_test_scaled)

# Calculate Mean Squared Error (MSE) on test set
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error (MSE) on Test Set:", mse)
```
